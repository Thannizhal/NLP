```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("We just opened our wings, the flying part is coming soon")
```

```python
for token in doc:
    print(token.text)
```

```python
for token in doc:
    if token.is_stop:
        print(token.text)
```

```python
def preprocess(text):
    doc = nlp(text)
    not_stopwords = [token.text for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(not_stopwords)
```

```python
preprocess("We just opened our wings, the flying part is coming soon")
```

```python
preprocess("Musk wants time to prepare a trail over his")
```
